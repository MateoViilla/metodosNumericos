```python
#
#Bisección
#

def biseccion():
    
    x = Symbol('x') #se declara la variable simbólica de la función
    fun = vars(math) #nos aseguramos que podamos utilizar funciones trigonométricas u otras funciones especiales
    umbral= 0.00001 #Definimos el umbral del error igual a 10^-5
    funcion = input('Ingrese la función o ecuación a solucionar en función de x: ') #Ingreso de la función por parte del usuario
    punto_inicial = float(input('Ingrese el punto inicial: ')) #Ingreso de punto inicial por parte del usuario
    punto_final = float(input('Ingrese el punto final: ')) #Se pide el punto final
   
    Fxi = eval(funcion,fun,{'x':punto_inicial}) #Valor de la función ingresada en el punto inicial recien ingresado
    Fxf = eval(funcion,fun,{'x':punto_final}) #Valor de la función ingresada en el punto final recien ingresado

# si el valor absuluto de la funcion evaluada en el punto inicial  es menor o igual que el umbral o la misma en el punto final entonces se puede tomar como valor de la raíz y dar finalidad al método
    if(abs(Fxi)<= umbral):
        raiz= punto_inicial
        print('El valor de la raiz es:' , raiz)
    elif(abs(Fxf)<= umbral):
        raiz= punto_final
        print('El valor de la raiz es:' , raiz)

# Este método consiste en realizar una busqueda más rápida al reducir los intervalos cada vez en su mitad y 
# hacer las evaluaciones de las funciones 
    punto_nuevo = (punto_inicial + punto_final)/2


#Se crea el bucle que analice el resultado de las funciones y en cualquiera de los casos se cambian el rango de busqueda en su mitad 
#para la nueva iteración hasta que sea encontrada la raíz de igual manera se establece un contador para tener un limite de iteraciones 

    cont = 0
    while(abs(eval(funcion,fun,{'x':punto_nuevo})) >= umbral and cont <= 30):
        if(eval(funcion,fun,{'x':punto_inicial}) * eval(funcion,fun,{'x':punto_nuevo}) <= 0):
            punto_final= punto_nuevo
        else:
            punto_inicial = punto_nuevo

        punto_nuevo = (punto_final + punto_inicial)/2
        
        cont = cont + 1

    if(abs(eval(funcion,fun,{'x':punto_nuevo})) <= umbral):
        raiz = punto_nuevo
        print('Raíz : ' , raiz)
    else:
        print('No se encontró la raíz')
```

```python
# Metodo de deteccion por cruces de busqueda Incremental
def busquedaIncremental(funcion,puntoInicial,incremento,error,iteraciones):
    fun = vars(math) 
    
    if abs(eval(funcion,fun,{'x': puntoInicial})) <= error:
        return puntoInicial #retorna el punto inicial si es solucion

    puntoNuevo = puntoInicial + incremento # calcula xn segun el metodo
    
    #itera si f(xi) y f(xn) tienen signos iguales o alcanza el maximo de iteraciones
    while eval(funcion,fun,{'x': puntoInicial})*eval(funcion,fun,{'x': puntoNuevo}) > 0 and iteraciones > 0:
        
        puntoInicial = puntoNuevo # aumenta el punto inicial
        puntoNuevo = puntoInicial + incremento # calcula xn segun el metodo
        
        iteraciones -= 1 # reduce en 1 el numero de iteraciones
    
    # mensaje si no enceuntra solucion en las x iteraciones
    if iteraciones == 0:
        return('no se encontro la raiz en el numero maximo de iteraciones')
    
    return puntoNuevo #retorna xn como solucion a f(x)
```
